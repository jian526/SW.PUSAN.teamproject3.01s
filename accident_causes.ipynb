```python
import pandas as pd

# train.csv 읽어오기
df1 = pd.read_csv('./data/train.csv')

df1.head()
```

```python
# 필요 없는 열 제거
df2 = df1.drop(['ID','사고인지 시간', '습도', '연면적', '층 정보', '물적사고', '공종', '사고객체', '작업프로세스','장소', '부위'], axis=1)
df2.head()
```

```python
df2['공사종류'] = df2['공사종류'].str.split('/').str[0].str.strip()
df2.head()
```

```python
df2['인적사고'] = df2['인적사고'].str.split('(').str[0].str.strip()
df2.head()
```

```python
df2['날씨'].unique()
```

```python
df2['발생일시'] = df2['발생일시'].str.split(' ').str[0].str.strip()
df2.head()
```

```python
df2.columns = [
    'accident_date',
    'weather',
    'temperature',
    'construction_type',
    'accident_type',
    'cause',
    'prevention'
]
df2.head()
```

```python
df2.info()
```

```python
import pandas as pd
import mysql.connector

#  1. 날짜 전처리 (시간 제외하고 날짜만)
df2['accident_date'] = pd.to_datetime(df2['accident_date'], errors='coerce').dt.date

# 2. 결측값 제거 + 튜플로 변환
rows = [tuple(pd.Series(row).fillna('')) for _, row in df2.iterrows()]
batch_size = 500  # 원하는 배치 크기 설정

# 3. MySQL 연결
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345",         
    database="accident_db"
)
cursor = conn.cursor()

# 4. 배치 단위로 INSERT
for i in range(0, len(rows), batch_size):
    batch = rows[i:i + batch_size]
    cursor.executemany("""
        INSERT INTO accident_cases (
            accident_date, weather, temperature,
            construction_type, accident_type, cause, prevention
        ) VALUES (%s, %s, %s, %s, %s, %s, %s)
    """, batch)
    conn.commit()
    print(f" {i + len(batch)} / {len(rows)} rows inserted")

# 5. 마무리
cursor.close()
conn.close()
print("🎉 모든 배치 저장 완료!")
```
